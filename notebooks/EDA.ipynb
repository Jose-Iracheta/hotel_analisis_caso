Perfilar datos maestros CATALOGO

```python
#Funcion para perfilar

def perfilar(df, nombre):
    df_perfilado = pd.DataFrame({
        "Tipo de Dato": df.dtypes,
        "Registros Totales": len(df),
        "Nulos": df.isnull().sum(),
        "Únicos": df.nunique(),
        "Duplicados": [df[col].duplicated().sum() for col in df.columns],
        "Vacíos": [df[col].apply(lambda x: x == '' or x == ' ').sum() if df[col].dtype == 'object' else 0 for col in df.columns]
    })

    print(f"---------- Perfilado de {nombre} ----------\n")
    display(df_perfilado)
```

```python
perfilar(df_empresas_raw, "EMPRESAS")
```

```python
perfilar(df_agencias_raw, "AGENCIAS")

#mostrar duplicados por columna
df_agencias_raw[df_agencias_raw.duplicated(subset='Agencia_nombre', keep=False)]
#eliminamos duplicados, dejamos primer registro y guardamos nuevo df para comparar
df_agencias=df_agencias_raw.drop_duplicates(subset='Agencia_nombre', keep='first')
#imprimir datos
perfilar(df_agencias, "AGENCIAS")
```

```python
perfilar(df_segmentos_raw, "SEGMENTOS")
```

Perfilar tablas transaccionales

```python
def perfilar_transaccional(df, nombre):
    print(f"\n----- Perfilado de {nombre} -----\n")
    
    # Información general
    print("Registros totales:", len(df))
    print("Columnas:", df.shape[1])
    print("Filas duplicadas:", df.duplicated().sum())
    print("Nulos totales:", df.isnull().sum().sum())
    
    # Perfil por columna
    print("\n--- Perfil por columna ---")
    perfil = pd.DataFrame({
        "Tipo de dato": df.dtypes,
        "Registros Totales": len(df),
        "Nulos": df.isnull().sum(),
        "Duplicados": [df[col].duplicated().sum() for col in df.columns],
        "Únicos": df.nunique(),
        "Vacíos": [df[col].apply(lambda x: x == '' or x == ' ').sum() if df[col].dtype == 'object' else 0 for col in df.columns]
    })
    display(perfil)

```

```python
perfilar_transaccional(df_reservas_raw, "RESERVAS")
```

INGRESOS POR EMPRESA

```python
df_analisis = df_ocupaciones_raw.merge(df_agencias, on='ID_Agencia', how='left') 
```

```python
rentabilidad_agencia = (
    df_analisis.groupby('Agencia_nombre')['ingresos_habitaciones']
    .sum()
    .reset_index()
    .rename(columns={'ingresos_habitaciones': 'Ingresos_totales'})
    .sort_values('Ingresos_totales', ascending=False) 
    .head(5)
)

```

```python
# Escalar ingresos a millones
rentabilidad_agencia['Ingresos_millones'] = rentabilidad_agencia['Ingresos_totales'] / 1_000_000

colores = plt.cm.tab10.colors[:len(rentabilidad_agencia)]

# --- Graficar ---
plt.figure(figsize=(12,6))
barras = plt.bar(
    rentabilidad_agencia['Agencia_nombre'],
    rentabilidad_agencia['Ingresos_millones'],
    color=colores
)

# Título y etiquetas
plt.title('Top 5 Agencias por Ingresos', fontsize=14, fontweight='bold')
plt.xlabel('Agencia')
plt.ylabel('Ingresos Totales (Millones MXN)')

# Nombres del eje X verticales
plt.xticks(rotation=90, ha='center')

# Mostrar valor exacto sobre cada barra
for barra, valor_real in zip(barras, rentabilidad_agencia['Ingresos_totales']):
    plt.text(
        barra.get_x() + barra.get_width() / 2,
        barra.get_height(),
        f"${valor_real:,.0f}",  # separador de miles
        ha='center',
        va='bottom',
        fontsize=10,
        fontweight='bold'
    )

plt.tight_layout()
plt.show()
```

EMPRESAS CON MAS RESERVAS

```python
df_reservas_hoteles = df_reservas_raw.merge(df_empresas_raw, on='ID_empresa', how='left' )

Hoteles_reservaciones = (
    df_reservas_hoteles.groupby('Empresa_nombre')['ID_Reserva']
    .count()
    .reset_index()
    .rename(columns={'ID_Reserva': 'Total_Reservas'})
    .sort_values('Total_Reservas', ascending=False)
)
```

```python
plt.figure(figsize=(10,5))
plt.bar(Hoteles_reservaciones['Empresa_nombre'][:7], Hoteles_reservaciones['Total_Reservas'][:7])
plt.title('Empresas con más Reservas')
plt.xlabel('Empresa')
plt.ylabel('Total de Reservas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

for x, y in zip(Hoteles_reservaciones['Empresa_nombre'][:7], Hoteles_reservaciones['Total_Reservas'][:7]):
    plt.text(x, y, f"{y:,}", ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.show()
```

TENDENCIAS DE INGRESOS Y CRECIMIENTO

```python
df_analisis['Año'] = df_analisis['Fecha_hoy'].dt.year

# Agrupar ingresos por año
ingresos_por_año = (
    df_analisis.groupby('Año')['ingresos_habitaciones']
    .sum()
    .reset_index()
    .rename(columns={'ingresos_habitaciones': 'Ingresos'})
)

# Escalar ingresos a millones
ingresos_por_año['Ingresos_millones'] = ingresos_por_año['Ingresos'] / 1_000_000

# Calcular % de crecimiento
ingresos_por_año['Crecimiento_%'] = ingresos_por_año['Ingresos'].pct_change() * 100
ingresos_por_año['Crecimiento_%'] = ingresos_por_año['Crecimiento_%'].fillna(0)

# --- Graficar ---
plt.figure(figsize=(10,5))

# Línea de ingresos
plt.plot(
    ingresos_por_año['Año'],
    ingresos_por_año['Ingresos_millones'],
    marker='o',
    linestyle='-',
    color='skyblue',
    linewidth=2,
    label='Ingresos (Millones MXN)'
)

# Línea de % de crecimiento
# Escalar crecimiento para visualizar en mismo eje
escala = ingresos_por_año['Ingresos_millones'].max() / ingresos_por_año['Crecimiento_%'].max()
plt.plot(
    ingresos_por_año['Año'],
    ingresos_por_año['Crecimiento_%'] * escala,
    marker='s',
    linestyle='--',
    color='orange',
    linewidth=2,
    label='% Crecimiento (escalado)'
)

# Etiquetas de ingresos
for x, y in zip(ingresos_por_año['Año'], ingresos_por_año['Ingresos_millones']):
    plt.text(x, y, f"{y:,.1f}", ha='center', va='bottom', fontsize=10, fontweight='bold')

# Etiquetas de crecimiento
for x, y in zip(ingresos_por_año['Año'], ingresos_por_año['Crecimiento_%'] * escala):
    plt.text(x, y, f"{y/escala:.1f}%", ha='center', va='bottom', fontsize=10, color='orange', fontweight='bold')

plt.title('Tendencia de Ingresos y % de Crecimiento por Año', fontsize=14, fontweight='bold')
plt.xlabel('Año')
plt.ylabel('Ingresos (Millones MXN)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(ingresos_por_año['Año'])
plt.legend()
plt.tight_layout()
plt.show()
```
